```python
import numpy as np
CL0 = 0
CLalpha = 5.7
CD0 = 0.02
k = 0.01

def get_CL(CL0,CLalpha, alpha):
    return CL0 + CLalpha*alpha

def get_CD(CD0,CL, k):
    return CD0 + (CL**2)*k



def tension_lines(alpha,alpha_0,theta_LE, theta_TE):
    CL = get_CL(CL0,CLalpha,alpha)
    CD = get_CD(CD0,CL,k)
    tension_TE = 0.75*(CL*np.cos(theta_TE-alpha+alpha_0)+CD*np.sin(theta_TE-alpha+alpha_0))
    tension_LE = 0.75*(CL*np.cos(theta_LE+alpha-alpha_0)+CD*np.sin(theta_LE+alpha-alpha_0))
    return tension_TE, tension_LE

alpha_0 = np.radians(0)
theta_LE = np.radians(10)
theta_TE = np.radians(10)
alphas = np.radians(np.linspace(0,10,100))

ratio = []
tension_LE = []
tension_TE = []
for alpha in alphas:
    a,b = tension_lines(alpha,alpha_0,theta_LE, theta_TE)
    tension_LE.append(a)
    tension_TE.append(b)
    ratio.append(b/a)

import matplotlib.pyplot as plt
plt.plot(np.degrees(alphas),ratio)
plt.xlabel('alpha [deg]')
plt.ylabel('tension_LE/tension_TE')

plt.figure()
plt.plot(np.degrees(alphas),tension_LE)
plt.plot(np.degrees(alphas),tension_TE)
```
