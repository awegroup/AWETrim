```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate lift and drag coefficients
def aero_coeffs_elliptical(alpha, AR, e=0.9):
    CL = 2 * np.pi * np.sin(np.radians(alpha))  # Simplified lift coefficient
    CD = 0.01 + CL**2 / (AR * e * np.pi)  # Simplified drag coefficient
    return CL, CD

# Parameters
alpha_range = np.linspace(0, 20, 100)  # Angle of attack range [-10°, 20°]
AR = 10  # Aspect ratio
oswald_efficiencies = [0.7, 0.85, 1.0]  # Different Oswald efficiency factors

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

for e in oswald_efficiencies:
    CL, CD = aero_coeffs_elliptical(alpha_range, AR, e=e)
    
    # Plot Lift Coefficient vs. Alpha
    axs[0, 0].plot(alpha_range, CL, label=f'e = {e:.2f}')
    axs[0, 0].set_xlabel('Angle of Attack (α) [deg]')
    axs[0, 0].set_ylabel('Lift Coefficient ($C_L$)')
    axs[0, 0].set_title('Lift Coefficient vs. Angle of Attack')
    axs[0, 0].grid()
    
    # Plot Drag Coefficient vs. Alpha
    axs[0, 1].plot(alpha_range, CD, label=f'e = {e:.2f}')
    axs[0, 1].set_xlabel('Angle of Attack (α) [deg]')
    axs[0, 1].set_ylabel('Drag Coefficient ($C_D$)')
    axs[0, 1].set_title('Drag Coefficient vs. Angle of Attack')
    axs[0, 1].grid()
    
    # Plot CL/CD
    axs[1, 0].plot(alpha_range, CL / CD, label=f'e = {e:.2f}')
    axs[1, 0].set_xlabel('Angle of Attack (α) [deg]')
    axs[1, 0].set_ylabel('$C_L / C_D$')
    axs[1, 0].set_title('Efficiency ($C_L / C_D$) vs. Angle of Attack')
    axs[1, 0].grid()
    
    # Plot CL^3/CD^2
    axs[1, 1].plot(alpha_range, CL**3 / CD**2, label=f'e = {e:.2f}')
    axs[1, 1].set_xlabel('Angle of Attack (α) [deg]')
    axs[1, 1].set_ylabel('$C_L^3 / C_D^2$')
    axs[1, 1].set_title('Power Efficiency ($C_L^3 / C_D^2$) vs. Angle of Attack')
    axs[1, 1].grid()

# Add legends and layout adjustments
for ax in axs.flat:
    ax.legend()
plt.tight_layout()
plt.show()
```
